## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from IPython.display import display

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

# Display the data table for preview
```

```python
print(f"Number of rows and columns: {mouse_metadata.shape}\n")
print(f"{mouse_metadata.dtypes}\n")
print(f"NaN values exist: {mouse_metadata.isnull().values.any()}\n")
display(mouse_metadata.head())
display(mouse_metadata.describe())
print(f"All values of 'Mouse ID' are unique: {mouse_metadata['Mouse ID'].is_unique}\n")
print(f"{mouse_metadata['Drug Regimen'].value_counts()}\n")
print(f"{mouse_metadata['Sex'].value_counts()}\n")
```

```python
print(f"Number of rows and columns: {study_results.shape}\n")
print(f"{study_results.dtypes}\n")
print(f"NaN values exist: {study_results.isnull().values.any()}\n")
display(study_results.head())
display(study_results.describe())
print(f"Number of unique mice: {study_results['Mouse ID'].nunique()}\n")
print(f"{study_results['Mouse ID'].value_counts()}\n")
print(f"{study_results['Timepoint'].value_counts()}\n")
print(f"{study_results['Metastatic Sites'].value_counts()}\n")
```

```python
# Testing whether all 'Mouse ID' values match between the two tables
s_MouseID_metadata = mouse_metadata['Mouse ID'].copy().sort_values().reset_index(drop=True)
s_MouseID_results = study_results['Mouse ID'].copy().drop_duplicates().sort_values().reset_index(drop=True)
test = s_MouseID_metadata.equals(s_MouseID_results)
print(f"'Mouse ID' values match between the two tables: {test}\n")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dups = study_results[study_results.duplicated(['Mouse ID', 'Timepoint'], keep=False)].copy()
display(dups)

# Show all duplicated 'Mouse ID'
print("Duplicate 'Mouse ID' report:")
print(dups['Mouse ID'].value_counts())
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
droprows = study_results.loc[study_results['Mouse ID'] == 'g989']
display(droprows)
drop_count = len(droprows)
```

```python
# Create clean DataFrames by dropping the duplicate mouse by its ID.

# Merge DataFrames
alldata = pd.merge(study_results, mouse_metadata, on="Mouse ID")


alldata_cl = alldata.loc[alldata['Mouse ID'] != 'g989'].reset_index(drop=True)
display(alldata_cl.head())
```

```python
# Checking the number of mice in the clean DataFrame.
print(f"The number of mice in the cleaned combined dataset is: {alldata_cl['Mouse ID'].nunique()}\n")
print(f"The number of records in the cleaned combined dataset is {len(alldata_cl)}")
print(f"The number of records in the original Study Results dataset is {len(study_results)}, and \
{drop_count} rows were dropped.\n")
if len(study_results) - drop_count == len(alldata_cl):
    print("Data cleaning successful.")
else:
    print("There was a problem with the data cleaning process.")
```
