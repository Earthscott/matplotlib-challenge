# Observations and Inferences

1. The number of mice tested with each drug regimen is essentially equivalent (n = 24 or 25), so variation in the initial population of drug regimen replicates is not a source of bias. However, a supplemental analysis demonstrates that the number of mice in each drug regimen with reported results at day 45 is considerably less, ranging from 7 to 21. As a result, variations in drug regimen replicates across the study duration may introduce bias and complicate comparison of results.
2. The distribution of sex is essential equivalent (male = 125, female = 123). However, a supplemental analysis shows that sex is not distributed equally for many drug regimens. Statistical analyses should be conducted to determine whether results are influenced by mouse sex.
3. The range and distribution of final tumor volume for Capomulin is comparable to Ramicane, while the range of the final tumor volume for other eight drug regimens is much higher, as shown in the supplemental analysis. Further analysis could be conducted to determine whether Capomulin and Ramicane differ in other meaningful ways.

## Exploratory Analysis of Capomulin Performance

### Boxplots of Final Tumor Volume for Capomulin and Three Other Drug Regimens

```python
# Select 'Drug Regimens' and filter master df
drug_subset = 'Capomulin Ramicane Infubinol Ceftamin'.split()
alldata_subset = alldata[alldata['Drug Regimen'].isin(drug_subset)].copy().reset_index()
alldata_subset

# Get the last timepoint for each mouse
df_Subset_EndPt = alldata_subset.groupby(['Drug Regimen', 'Mouse ID'])['Timepoint'].max().reset_index()

# Merge group df with the master df to get the tumor volume at the last timepoint
merge_cols = df_Subset_EndPt.columns.to_list()
df_Subset_EndPt = df_Subset_EndPt.merge(alldata_subset[merge_cols + ['Tumor Volume (mm3)']], on=merge_cols, how='left')

# Calculate quartiles, IQR, identify potential outliers, and prepare boxplot list of lists
box_data = []
for drug in drug_subset:
    s_tumvol = df_Subset_EndPt.loc[df_Subset_EndPt['Drug Regimen'] == drug]['Tumor Volume (mm3)'].copy()
    lowerq = round(s_tumvol.quantile(0.25),6)
    upperq = round(s_tumvol.quantile(0.75),6)
    iqr = round(upperq - lowerq,6)
    print(f"{drug}: quartiles = {lowerq}, {upperq}  IQR = {iqr}")
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    s_lowout = list(s_tumvol[s_tumvol < lower_bound])
    s_highout = list(s_tumvol[s_tumvol > upper_bound])
    if len(s_lowout) == 0:
        print("No outliers below the lower bound were identified")
    else:
        print(f"*** Potential outlier(s) below the lower bound of {lower_bound} were identified: {', '.join([str(i) for i in s_lowout])}")
    if len(s_highout) == 0:
        print("No outliers above the upper bound were identified")
    else:
        print(f"*** Potential outlier(s) above the upper bound of {upper_bound} were identified: {', '.join([str(i) for i in s_highout])}")
    print()
    box_data.append(list(s_tumvol))
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
medianprops = dict(linewidth=1, color='black')
flierprops = dict(markerfacecolor='lightgrey', markersize=10)
bplot = ax1.boxplot(box_data, medianprops=medianprops, flierprops=flierprops, patch_artist=True)
colors = ['pink', 'lightblue', 'lightgreen', 'lightyellow']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
ax1.set_title('Boxplots of Final Tumor Volume by Drug Regimen', fontsize=14, pad=12)
ax1.set_ylabel('Tumor Volume ($mm^3$)', fontsize=11)
ax1.set_xticklabels(drug_subset, fontsize=12)
plt.show()
```

### Scatter Plot of Mean Tumor Volume Versus Mouse Weight for Mice Treated With Capomulin

```python
# Generate a scatter plot of mean tumor volume vs. mouse weight for the Capomulin regimen
alldata_capo = alldata.loc[alldata['Drug Regimen'] == 'Capomulin'].copy()
df_wt_tvol = alldata_capo.groupby(['Mouse ID','Weight (g)'])['Tumor Volume (mm3)'].mean().reset_index()
params3 = {'xlabel': 'Mouse Weight (g)',
          'ylabel': 'Mean Tumor Volume ($mm^3$)',
          'ylim': (32,48),
          'xlim': (14,26),
          'grid': True,
          's': 30}
fig3, ax3 = plt.subplots()
ax3 = plt.gca()
ax3.set_title('Mean Tumor Volume versus Mouse Weight\nfor Mice Treated With Capomulin', fontsize=14, pad=20)
df_wt_tvol.plot(kind='scatter', x='Weight (g)', y='Tumor Volume (mm3)', **params3, ax=ax3);
```

### Correlation Between Mean Tumor Volume and Mouse Weight for Mice Treated With Capomulin

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
x_values = df_wt_tvol['Weight (g)'].values
y_values = df_wt_tvol['Tumor Volume (mm3)'].values

# Correlation coefficient
correlation = st.pearsonr(y_values,x_values)
print(f"The correlation between both factors is {round(correlation[0],3)}")
```

### Regression Model for Mean Tumor Volume on Mouse Weight for Mice Treated With Capomulin

```python
# Linear regression model
(slope, intercept, r, p, se) = st.linregress(x_values, y_values)

regress_values = x_values * slope + intercept
eqtn = "y = " + str(round(slope,3)) + "x + " + str(round(intercept,3))

ax3.plot(x_values, regress_values, 'r-')
ax3.annotate(eqtn, (15,46), color='r', fontsize='x-large')
ax3.annotate(f"R$^2$ = {r*r:.3f}", (15,44.5), color='r', fontsize='x-large')
fig3
```
